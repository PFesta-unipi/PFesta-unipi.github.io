```julia

````julia
import Pkg; Pkg.add("Mods")
# caricare le librerie
using Mods
using Plots
````

````julia
# Creiamo elementi nella classe di resto modulo 10
a = mod(7, 10)
b = mod(5, 10)
c = mod(-1, 10)

# Operazioni elementari
println(mod(a + b,10))
println(mod(a * b, 10))
println( c+1 )
````

Verifichiamo sperimentalmente che

$$
(a+b) \text{mod } n=((a \text{ mod }n)+(b \text{ mod } n))\text{mod } n
$$

````julia
# assegno dei valori ad a, b, n
a, b, n = 7, 9, 5

left = mod(a + b, n)
right = mod(mod(a, n) + mod(b, n), n)

left == right
````

Per calcolare il reciproco modulo $n$, si usa il comando `invmod`:

```julia
a = invmod(3, 7)
println("a=", a)
println(mod(a*3,7))
println("1/a =", mod(a*3,7))
```

Il comando `invmod` restituisce errore se gli argomenti non sono primi tra loro

````julia
a = invmod(3, 7)
println("a = ", a)
println("1/a = ", mod(a*3,7))
````

# 2. Calcolo simbolico

Mentre finora abbiamo usato Julia come una "calcolatrice numerica", il pacchetto `Symbolics` permette di manipolare espressioni matematiche senza valutarle numericamente.

Per definire variabili simboliche si usa la macro `@variables`:

```julia
# definiamo le variabili simboliche x e y
@variables x y

# definiamo una espressione simbolica in funzione di x e y
expr = (x + y)^2

# La funzione expand restituisce un'espansione dell'espressione, cioe' x^2 + 2x*y + y^2
espansa = expand(expr)
println(espansa)


````julia
import Pkg; Pkg.add("Symbolics")
using Symbolics
````

````julia
# definiamo le variabili simboliche x e y
@variables x y

# definiamo una espressione simbolica in funzione di x e y
expr = (x + y)^2

# La funzione expand restituisce un'espansione dell'espressione, cioe' x^2 + 2x*y + y^2
espansa = expand(expr)
println(espansa)
````

Possiamo valutare l'espressione `expr` in opportuni valori di `x` e `y`:

```julia
substitute(expr, x => 3)
```

oppure derivare:

```julia
f = x^2 + cos(x)
D = Differential(x)
df = expand_derivatives(D(f))
```

Le espressioni simboliche possono essere trasformate in `function` mediante il comando `build_function`, come ad esempio nelle celle seguenti.

````julia
substitute(expr, x => 3)
````

````julia
@variables x
f = x^2 + cos(x)
D = Differential(x)
# df e' l'espressione simbolica della derivata di f
df = expand_derivatives(D(f))
println("f' = ", df)

# df_expr e' la funzione della variabile x, ottenuta trasformando l'espressione simbolica `df` in `function` della variabile x
df_expr = build_function(df, x, expression=Val{false})
x0 = 0.0
df_x0 = df_expr(x0)
println("f'($x0) = ", df_x0)
````

**Esercizio 6.**  Definire $x$ come variabile simbolica e creare l'espressione di $f(x) = \frac{x^2 - 1}{x^2 + 1}$. Calcolare la derivata $f'(x)$ simbolicamente. Trovare l'equazione della retta tangente al grafico di $f$ nel punto $x_0 = 1$.

**Suggerimento:** Usare la funzione `substitute` per valutare le espressioni nel punto $x_0$ nell'equazione è $y = f(x_0) + f'(x_0)(x - x_0)$

````julia
@variables x
f = (x^2 - 1) / (x^2 + 1)
D=Differential(x)
df=expand_derivatives(D(f))
tangente= substitute(f, x=>1) + substitute(df, x=>1)*(x-1)
println(df)
println(tangente)
````

**Esercizio 7.** Data $g(x) = e^x \cos(x)$, calcolare le prime quattro derivate in $x=0$. Costruire il polinomio di Taylor di quarto grado
$$P_4(x) = \sum_{n=0}^{4} \frac{g^{(n)}(0)}{n!} x^n.$$
Tracciare sullo stesso grafico la funzione originale $g(x)$ e il suo polinomio di Taylor $P_4(x)$ nell'intervallo $[-2, 2]$.

**Suggerimento.** Per calcolare i coefficienti trasformare le espressioni simboliche delle derivate in `function` mediante il comando `build_function`, quindi i coefficienti saranno delle variabili `Float64`. Per tracciare il grafico trasformare la funzione originale e il polinomio in `function`, sempre con il comando `build_function`.

````julia
@variables x
g = exp(x)*cos(x)
D=Differential(x)
#g1=expand_derivatives(D(g))
#g2=expand_derivatives(D(g1))
#g3=expand_derivatives(D(g2))
#g4=expand_derivatives(D(g3))
P = substitute(g, x=>0)
d=expand_derivatives(D(g))
for i in 1:4
  P=P+substitute(d, x=>0)*x^i
  d=expand_derivatives(D(d))/(i+1)
end
println(P)
g_expr=build_function(g, x, expression=Val{false})
P_expr=build_function(P, x, expression=Val{false})
t=range(-2.0, stop=2.0, length=100)
y1=g_expr.(t)
y2=P_expr.(t)
plot(t, y1)
plot!(t, y2)
````
