# 1. Superfici e curve di livello

Vogliamo disegnare la funzione $f(x,y)=x^2+y^2$ definita sul rettangolo $[-0.5,0.5]\times [-0.5,0.5]$, e le corrispondenti curve di livello.

Carichiamo prima le librerie grafiche
```julia
# caricare le librerie
using Plots
# gr backend for 3D interaction
gr()
```

Poi definiamo la funzione, l'intervallo ``discretizzato'' e usiamo la funzione `surface`, a cui passiamo direttamente la funzione:
```julia
# f(x,y) = x^2 + y^2
f(x, y) = x^2 + y^2
x = -0.5:0.01:0.5
y = -0.5:0.01:0.5
# In Julia, you can pass the function and ranges directly
p1 = surface(x, y, f, title="f(x,y)", xlabel="x", ylabel="y")
```
Per disegnare le curve di livello, usiamo con la stessa sintassi la funzione `contour`
```julia
p2 = contour(x, y, f, title="curve di livello", xlabel="x", ylabel="y",aspect_ratio=:equal)
plot(p1, p2, layout=(1,2))

# 2. Curve parametriche

Una curva parametrica in $\mathbb{R}^3$ è definita da tre equazioni che descrivono la dipendenza di ciascuna coordinata $x$, $y$, $z$ dal parametro $t$. Consideriamo la curva di equazioni
\begin{eqnarray*}
&& x(t)=(1+t^2)\cos(20t),\\
&& y(t)=(1+t^2)\sin(20t),\\
&& z(t)=t,
\end{eqnarray*}
con $t\in [-5,5]$. Per disegnarla possiamo usare i comandi

```julia
t = -5:0.01:5
x = (1.0.+t.^2).*cos.(20.0.*t)
y = (1.0.+t.^2).*sin.(20.0.*t)
z = t
plot(x, y, z, title="curva in R3")

````julia
t = -5:0.01:5
x = (1.0.+t.^2).*cos.(20.0.*t)
y = (1.0.+t.^2).*sin.(20.0.*t)
z = t
plot(x, y, z, title="curva in R3")
````

# 3. Superfici parametriche

Una superficie parametrica in $\mathbb{R}^3$ è definita da tre equazioni che descrivono la dipendenza di ciascuna coordinata $x$, $y$, $z$ da due parametri $r$, $t$.

Vogliamo disegnare la superficie in $\mathbb{R}^3$ definita dalle equazioni parametriche
\begin{eqnarray*}
&& x(r,t)=r\cos(t),\\
&& y(r,t)=r\sin(t),\\
&& z(r,t)=r,
\end{eqnarray*}
con $r\in[0,1]$ e $t\in[0,2\pi]$.

```julia
# Definiamo i range dei parametri
r_vals = range(0, 1, length=50)
t_vals = range(0, 2π, length=50)

# Generiamo le matrici 2D con le coordinate
X = [r * cos(t) for t in t_vals, r in r_vals]
Y = [r * sin(t) for t in t_vals, r in r_vals]
Z = [r for t in t_vals, r in r_vals]

# Plot della superficie parametrica
surface(X, Y, Z,
    aspect_ratio=:equal,
    title="Superficie",
    xlabel="x", ylabel="y", zlabel="z",
    color=:viridis)
```

Come potete osservare, otteniamo un cono (perché?).


````julia

# Definiamo i range dei parametri
r_vals = range(0, 1, length=50)
t_vals = range(0, 2π, length=50)

# Generiamo le matrici 2D con le coordinate
X = [r * cos(t) for t in t_vals, r in r_vals]
Y = [r * sin(t) for t in t_vals, r in r_vals]
Z = [r for t in t_vals, r in r_vals]

# Plot della superficie parametrica
surface(X, Y, Z,
    aspect_ratio=:equal,
    title="Superficie",
    xlabel="x", ylabel="y", zlabel="z",
    color=:viridis)
````
